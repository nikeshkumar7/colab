```python

import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img = cv2.imread('/content/God-Shiva-Neon-Colours-iPhone-Wallpaper-HD.jpg',0)
```

```python
plt.imshow(img)
plt.show()
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
erosion = cv2.erode(img,kernel,iterations = 1)
```

```python
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.imshow(dilation)
plt.show()
```
