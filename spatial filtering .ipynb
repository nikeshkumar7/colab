```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img_noisy1 = cv2.imread('/content/noisysalterpepper.png', 0)
```

```python
plt.imshow(img_noisy1)
plt.show()
```

```python
m, n = img_noisy1.shape
```

```python
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img_noisy1[i-1, j-1],
               img_noisy1[i-1, j],
               img_noisy1[i-1, j + 1],
               img_noisy1[i, j-1],
               img_noisy1[i, j],
               img_noisy1[i, j + 1],
               img_noisy1[i + 1, j-1],
               img_noisy1[i + 1, j],
               img_noisy1[i + 1, j + 1]]
```

```python
temp = sorted(temp)
```

```python
img_noisy1[i, j]= temp[4]
```

```python
img_noisy1 = img_noisy1.astype(np.uint8)
```
