```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img=cv2.imread('/content/1bit1.png')
img1=cv2.imread('/content/2bit2.png')
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img1)
plt.show()
```

```python
dest_and = cv2.bitwise_and(img, img1, mask = None)
```

```python
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
plt.imshow(dest_and)
plt.show()
```

```python
dest_or = cv2.bitwise_or(img, img1, mask = None)
```

```python
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
plt.imshow(dest_or)
plt.show()
```

```python
dest_xor = cv2.bitwise_xor(img, img1, mask = None)
```

```python
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
plt.imshow(dest_xor)
plt.show()
```

```python
dest_not1 = cv2.bitwise_not(img, mask = None)
dest_not2 = cv2.bitwise_not(img1, mask = None)
```

```python
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

```python
plt.imshow(dest_not1)
plt.show()
```

```python
plt.imshow(dest_not2)
plt.show()
```
