```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img=cv2.imread('/content/lord-rama-2.jpg')
plt.imshow(img)
plt.show()
```

```python
kernel1 = np.array([
    [1, 1, 1],
    [1, -8, 1],
    [1, 1, 1]
])
```

```python
kernel2 = np.array([
    [1, 1, 1],
    [1, -9, 1],
    [1, 1, 1]
])
```

```python
dst1 = cv2.filter2D(img, -1, kernel1)
dst2 = cv2.filter2D(img, -1, kernel2)
```

```python
sharpened_image = cv2.addWeighted(img, 1.5, dst1, -0.5, 0)
```

```python
plt.subplot(131), plt.imshow(img), plt.title('Original')
plt.subplot(132), plt.imshow(sharpened_image), plt.title('Sharpened')
plt.subplot(133), plt.imshow(dst1), plt.title('Laplacian Filtered')
```
